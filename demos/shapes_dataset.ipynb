# ``ShapesDataset`` demo.


``torch_tools`` comes with a synthetic dataset object that yields images containing different shapes, which are included with user-specified probabilities. The shapes have random RGB colours, as do the backgrounds.

## Examples

```python
from torch_tools import ShapesDataset


import matplotlib.pyplot as plt


data_set = ShapesDataset(
    square_prob=0.5,
    spot_prob=0.5,
    star_prob=0.5,
    octagon_prob=0.5,
    num_shapes=10,
)

figure, axes = plt.subplots(5, 5, figsize=(10, 10))

for idx, axis in enumerate(axes.ravel()):
    img, tgt = data_set[idx]

    axis.imshow(img.permute(1, 2, 0))
    axis.set_title(f"y = {tgt.numpy()}", fontsize=12)

    axis.set_xticks([])
    axis.set_yticks([])


figure.tight_layout(pad=0.1, h_pad=0.5, w_pad=0.0)
```
