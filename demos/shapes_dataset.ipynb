# ``ShapesDataset`` demo.


``torch_tools`` comes with a synthetic dataset object that yields images containing, spots, squares, both or neither. The images have white backgrounds, and each shape has randomly-choosen colours on $[0, 1)^{3}$.


## Examples

### 1. Spots only

```python
from torch_tools import ShapesDataset


import matplotlib.pyplot as plt


data_set = ShapesDataset(square_prob=0.0, spots_prob=0.5, num_spots=10)

figure, axes = plt.subplots(1, 8, figsize=(24, 5))

for idx, axis in enumerate(axes.ravel()):
    img, tgt = data_set[idx]

    axis.imshow(img.permute(1, 2, 0))
    axis.set_title(f"y = {tgt.numpy()}", fontsize=24)

    axis.set_xticks([])
    axis.set_yticks([])


figure.subplots_adjust(wspace=0.05)
```

### 2. Squares only

```python
import matplotlib.pyplot as plt


data_set = ShapesDataset(square_prob=0.5, spots_prob=0.0, num_squares=10)

figure, axes = plt.subplots(1, 8, figsize=(24, 5))

for idx, axis in enumerate(axes.ravel()):
    img, tgt = data_set[idx]

    axis.imshow(img.permute(1, 2, 0))
    axis.set_title(f"y = {tgt.numpy()}", fontsize=24)

    axis.set_xticks([])
    axis.set_yticks([])


figure.subplots_adjust(wspace=0.05)
```

### 3. Spots and squares

```python
import matplotlib.pyplot as plt


data_set = ShapesDataset(
    square_prob=0.5,
    spots_prob=0.5,
    num_squares=10,
    num_spots=10,
)

figure, axes = plt.subplots(1, 8, figsize=(24, 5))

for idx, axis in enumerate(axes.ravel()):
    img, tgt = data_set[idx]

    axis.imshow(img.permute(1, 2, 0))
    axis.set_title(f"y = {tgt.numpy()}", fontsize=24)

    axis.set_xticks([])
    axis.set_yticks([])


figure.subplots_adjust(wspace=0.05)
```
