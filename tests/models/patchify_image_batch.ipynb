# ``torch_tools.torch_utils.patchify_img_batch`` demo


First, we need a reference image.

```python
from skimage.data import horse
from skimage.transform import resize

import matplotlib.pyplot as plt

img = resize(horse(), (256, 256)).astype(float)

print(img.shape)


plt.imshow(img, cmap="Greys")
plt.show()


```

Now, we obtain the patches:

```python
from torchvision.transforms.functional import to_tensor
from torchvision.utils import make_grid

from torch_tools.torch_utils import patchify_img_batch

batch = to_tensor(img).unsqueeze(0)

patches = patchify_img_batch(batch, 32)

grid = make_grid(patches, nrow=8).permute(1, 2, 0)

plt.imshow(grid)
plt.show()
```
